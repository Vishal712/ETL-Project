```python
import pandas as pd
```

### Importing the trending data from ProjectData folder, and creating pandas dataframes from them.
Data Source: https://www.kaggle.com/datasnaek/youtube-new
Each csv file should contain the same columns, all Trending data from a different country.

```python
US_data = "ProjectData/USvideos.csv"
UK_data = "ProjectData/GBvideos.csv"
CA_data = "ProjectData/CAvideos.csv"
```

```python
us_videos = pd.read_csv(US_data)
uk_videos = pd.read_csv(UK_data)
ca_videos = pd.read_csv(CA_data)
```

Example of what columns are included for each Dataframe

```python
us_videos.head()
```

Cleaning the dataframes by dropping columns that we don't need. The same applied to all three dataframes. Printed a different country to show this.

```python
dropped_us = us_videos.drop(['trending_date','comments_disabled',
                             'ratings_disabled','video_error_or_removed', 'thumbnail_link'], axis = 1)
dropped_uk = uk_videos.drop(['trending_date','comments_disabled',
                             'ratings_disabled','video_error_or_removed', 'thumbnail_link'], axis = 1)
dropped_ca = ca_videos.drop(['trending_date','comments_disabled',
                             'ratings_disabled','video_error_or_removed', 'thumbnail_link'], axis = 1)
```

```python
dropped_uk.head()
```

### Imported list of categories text file to translate category_id to name of category, and stored as a dict.
Source of data from Github: https://gist.github.com/dgp/1b24bf2961521bd75d6c#gistcomment-2714395
stored as text file and transformed

```python
#Each line of text file is in form of # CategoryName. So we split the line by space, and 
#stored the number as a key and the CategoryName as its value.
category_list = open("ProjectData/categories.txt", "r").read()
category_list = category_list.split('\n')
category_translation = {}
for category in category_list:
    number = category.split(" ")[0]
    name = ''.join(category.split(" ")[1:])
    category_translation[int(number)] = name
```

```python
category_translation
```

Replaced the category number with a name

```python
dropped_us = dropped_us.replace({'category_id': category_translation })
dropped_uk = dropped_uk.replace({'category_id': category_translation })
dropped_ca = dropped_ca.replace({'category_id': category_translation })
```

```python
dropped_us.head()
```

Merge the tables together

```python
combined_videos = pd.merge(dropped_us, dropped_uk, how = 'inner', on = 'title')
combined_videos = pd.merge(combined_videos, dropped_ca, how = 'inner', on = 'title')
combined_videos = combined_videos[['title', 'channel_title_x', 'category_id_x', 'tags_x']]
combined_videos = combined_videos.drop_duplicates()
combined_videos = combined_videos.reset_index(drop = True)
```

```python
combined_videos.head()
```

Rename columns

```python
combined_videos = combined_videos.rename(columns={"channel_title_x": "channel", "category_id_x": "category", "tags_x": "tags"})
```

## This is now importing the top 500 YouTube channels in order to filter the list of shared videos further

The dataset can be found here: https://www.kaggle.com/balaka18/youtubers-popularity-dataset

```python
top_500 = "ProjectData/top_500.csv"
top_500_df = pd.read_csv(top_500)
```

```python
top_500_df.head()
```

Merged top 500 YouTube channels with our combined YouTube videos list with an inner merge

```python
combined_channels = pd.merge(top_500_df, combined_videos, how = 'inner', left_on = 'Ch_name', right_on = 'channel')
print(combined_channels.columns)
```

Reducing the columns included in final table for the channels in the top 500

```python
combined_channels = combined_channels.drop_duplicates()
combined_channels = combined_channels[["Rank", "Grade", "Ch_name", "Subscriptions", "Views", "category"]]
```

```python
combined_channels = combined_channels.drop_duplicates()
combined_channels = combined_channels.reset_index(drop = True)
```

```python
combined_channels
```

Reconfigured the rank column to represent integers instead of a string so someone can easily sort these if needed

```python
combined_channels_dict  = combined_channels.to_dict()
channel_ranks = combined_channels_dict["Rank"]
for index, rank in channel_ranks.items():
    rank_integer = int(rank.strip()[:-2])
    combined_channels_dict["Rank"][index] = rank_integer
```

```python
Channel_Ranks = pd.DataFrame.from_dict(combined_channels_dict)
```

```python
Channel_Ranks
```

Finally, sorted the table by its rank

```python
Channel_Ranks.sort_values(by=['Rank'])
```

Now uploading two created tables to Mongo DB. One containing Youtube Trending Videos in common with 3 countries, the
other with those channels in the top 500 of YouTube.

```python
import pymongo
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Declare the database
db = client.ETLYouTube_db
db.SharedTrendingVideos.drop()
db.SharedChannelsTop500.drop()

# Declare the collection
SharedTrendingVideos = db.SharedTrendingVideos
SharedChannelsTop500 = db.SharedChannelsTop500

#Convert our combined video dataframe into a dictionary and insert into Mongo
combined_vid_dict = combined_videos.to_dict('records')
SharedTrendingVideos.insert_many(combined_vid_dict)

#Convert our ranked combined channel dataframe into a dictionary and insert into Mongo.
ranked_vid_dict = Channel_Ranks.to_dict('records')
SharedChannelsTop500.insert_many(ranked_vid_dict)
```

Showing that we are able to get an entry in each collection using pymongo.
